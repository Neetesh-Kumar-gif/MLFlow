```python
import pandas as pd
import numpy as np
import os

import mlflow
import mlflow.sklearn

from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import ElasticNet

from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error, accuracy_score , roc_auc_score
from sklearn.model_selection import train_test_split

import argparse
```

```python
data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv',sep=";")
```

```python
data.head()
```

```python
train , test = train_test_split(data)
```

```python
X_train = train.drop(["quality"], axis = 1)
X_test = test.drop(["quality"], axis = 1)

y_train = train[["quality"]]
y_test = test[["quality"]]
```

```python
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
pred = rf.predict(X_test)
```

```python
pred_prob = rf.predict_proba(X_test)
pred_prob
```

```python
from sklearn.preprocessing import label_binarize

# Get the unique classes from training
classes = np.unique(y_test)

# Convert y_test to one-hot encoding
y_test_bin = label_binarize(y_test, classes=classes)

print("Shape of y_test_bin:", y_test_bin.shape)
print("Shape of pred_prob:", pred_prob.shape)
```

```python
roc_auc = roc_auc_score(y_test_bin, pred_prob, multi_class='ovr', average='macro')
print("ROC AUC Score:", roc_auc)
```
